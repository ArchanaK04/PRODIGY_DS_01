```python
import pandas as pd
#Using file1
df1 = pd.read_csv(r"API_SP.POP.TOTL_DS2_en_csv_v2_631214\API_SP.POP.TOTL_DS2_en_csv_v2_631214.csv")
df1
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Histogram for the population of a particular country over the years

country_name = 'Aruba'
df_country = df1[df1['Country Name'] == country_name]


df_country = df_country.drop(columns=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', 'Unnamed: 68'])

df_country = df_country.T
df_country.columns = ['Population']
df_country.index.name = 'Year'
df_country.reset_index(inplace=True)


# Plot histogram for the population over the years
plt.figure(figsize=(15, 6))
plt.hist(df_country['Year'], bins=len(df_country), weights=df_country['Population'], color='skyblue', edgecolor='black')
plt.xlabel('Year',fontsize=12, fontweight='bold')
plt.ylabel('Population',fontsize=12, fontweight='bold')
plt.title(f'Population of {country_name} Over the Years', fontsize=15, fontweight='bold')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability
plt.show()
```

```python
# Bar chart for the population among all countries in a particular year

# Choose a particular year
year = '2019'


df_year = df1[['Country Name', year]].dropna()


df_year_sorted = df_year.sort_values(by=year, ascending=False) 


plt.figure(figsize=(15, 10))
plt.bar(df_year_sorted['Country Name'], df_year_sorted[year], color='blue', edgecolor='black', alpha=0.7)


plt.xlabel('Country', fontsize=12, fontweight='bold')
plt.ylabel('Population', fontsize=12, fontweight='bold')
plt.title(f'Population Among Countries in {year}', fontsize=15, fontweight='bold')
plt.xticks(rotation=90, fontsize=3)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Using file2
df2 = pd.read_csv(r"API_SP.POP.TOTL_DS2_en_csv_v2_631214\Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_631214.csv")
df2
```

```python
# Number of Countries per region
df3=df2.groupby('Region').count()
df3
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df3.index,df3['Country Code'],color='Lightgreen',edgecolor='k',linestyle='--')
plt.xlabel('Regions', fontsize=12, fontweight='bold')
plt.ylabel('Number of Countries', fontsize=12, fontweight='bold')
plt.title('Number of Countries per Region', fontsize=15, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
#Number of Countries per income Group
df4=df2.groupby('IncomeGroup').count()
df4
```

```python
plt.figure(figsize=(6, 6))
plt.bar(df4.index,df4['Country Code'],color='Lightblue',edgecolor='k',linestyle=':')
plt.xlabel('IncomeGroup', fontsize=12, fontweight='bold')
plt.ylabel('Number of Countries', fontsize=12, fontweight='bold')
plt.title('Number of Countries per IncomeGroup', fontsize=15, fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```
